```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 5) # 5 points between 0 and 1
y = np.array([0, 1, 0, 1, 0])

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('1')
plt.show()
```

```python
omega = np.polyfit(x, y, 4) # 1元4次方程式
omega
```

```python
f = np.poly1d(omega)

plt.xlabel('x')
plt.ylabel('y')
plt.title('2')
plt.grid()
plt.scatter(x, y, marker='x', c='red')

plt.plot(x, f(x), color='green')

plt.show()
```
